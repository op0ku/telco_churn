```python
def perform_eda(df, columns=None, label='Churn'):
    """
    Perform exploratory data analysis (EDA) on the cleaned DataFrame.

    Parameters:
    - df (pd.DataFrame): cleaned DataFrame to perform EDA on.
    - columns (list): list of column names to perform EDA on. if None, all columns in the DataFrame
    - label: the response feature to be predicted

    Returns:
    EDA results and plots.
    """
    if columns == None:
        columns = df.columns.to_list()
    elif isinstance(columns, str):
        columns = [columns]
    
    for col in columns:
        #visualize percentage of churned customers
        if col == label:
            plt.pie(df[label].value_counts(),
                    labels=[i for i in df[label].value_counts().index],
                    autopct='%1.2f%%')
            plt.title(f'{label} Distribution')
            plt.show()

            #return percentage of churned customers
            churn_pct = round((df[label].value_counts()['Yes'] / len(df[label])) * 100, 2)
            no_churn_pct = round((df[label].value_counts()['No'] / len(df[label])) * 100, 2)
            print(f'In the last month, {no_churn_pct}% of customers continue to use the telco services. {churn_pct}% of customers churned.')

        #analyze distribution of numeric features and their relationships with the label(Churn) using histogram
        elif is_numeric_dtype(df[col]):
            sns.histplot(data=df, x=col, hue=label, multiple='stack', kde=True)
            plt.title(f'Distribution of {col} vs {label}')
            plt.xlabel(f'{col} values')
            plt.ylabel(f'Density')
            plt.show()
         
        #analyze distribution of category features using Pie chart
        else:
            plt.pie(df[col].value_counts(),
                    labels=[i for i in df[col].value_counts().index],
                    autopct='%1.2f%%')
            plt.title(f'{col} Distribution')
            plt.show()
            
            #analyse category features and their relationships with the label(Churn) using bar chart
            sns.countplot(data=df, x=col, hue=label)
            plt.title(f'Bar Chart: {col} vs {label}')
            plt.xlabel(f'{col}')
            plt.ylabel(f'count')
            plt.show()

            #return churn probabilities for the category features
            churn_probabilities = df.groupby(col)[label].value_counts(normalize=True) * 100
            print(f'Churn Probabilities of {col}:')
            for category_value in df[col].unique():
                churn_rate = churn_probabilities[category_value]['Yes'] if 'Yes' in churn_probabilities[category_value].index else 0
                print(f'A {category_value} customer has a churn probability of {round(churn_rate, 2)}%')
            print('\n')
```

# SHAP tutorial

```python
import matplotlib.pyplot as plt
import pandas as pd
import skops.io as sio
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from catboost import CatBoostClassifier
import shap
```

```python
from explore import wrangle, perform_eda #module to wrangle the dataset

#prepare the dataset
df = wrangle('C:/Users/solace.dark/Documents/Jupyter Notebook/Practical Projects/Customer Churn Prediction/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
#replace churn values with binary
df['Churn'] = df.loc[:, 'Churn'].replace({'No': 0, 'Yes': 1})
```

```python
#split dataset
X = df.drop(['Churn'], axis=1)
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.2, stratify=y, random_state=42
)
```

```python
#get indexes of category features
category_features = [i for i, col in enumerate(X_train.columns) if col in X_train.select_dtypes('object')]

#instantiate Catboost model
clf = CatBoostClassifier(iterations=150,
                        depth=8,
                        learning_rate=.07,
                        cat_features=category_features, 
                        scale_pos_weight=3,
                        random_state=42)

#train model on the dataset
clf.fit(X_train, y_train)
```

### explain model with SHAP

Shap will be used to explain how the catboot model made predictions on whether a customer churned.

```python
#get shap values
explainer = shap.TreeExplainer(clf)
train_shap_vals = explainer.shap_values(X_train)
test_shap_vals = explainer.shap_values(X_test)
```

```python
#visualize feature importances with beeswarm plot
shap.summary_plot(train_shap_vals, X_train, plot_type='dot')
```

This beeswarm plot shows the how important each feature was during churn prediction using the Catboost model. Features with wider range of shap values contributed the most to churn prediction.

From the plot the five most important features that contributed to  predicting churn using the Catboost model are Contract, Tenure, Monthly Charges, Internet Service and Payment Method

```python
# dependence plot for Contract
plt.figure(figsize=(8, 5))
shap.dependence_plot('Contract', test_shap_vals, X_test, interaction_index=None)

print('The Contract type that contributed most to churn prediction is month-to-month.')
```

```python
# dependence plot for Tenure
plt.figure(figsize=(8, 5))
shap.dependence_plot('tenure', test_shap_vals, X_test, interaction_index=None)

print('Tenure values increase as shap values decrease.\nThis indicates smaller tenure values contributed to churn predictions.')
```

```python
# dependence plot for MonthlyCharges
plt.figure(figsize=(8, 5))
shap.dependence_plot('MonthlyCharges', test_shap_vals, X_test, interaction_index=None)

print('Monthly charges increase with shap values.\nThis indicates higher monthly charges contributed to churn predictions.')
```

```python
# dependence plot for Internet Service
plt.figure(figsize=(8, 5))
shap.dependence_plot('InternetService', test_shap_vals, X_test, interaction_index=None)

print('The Internet Service type that contributed most to churn prediction is fibre optic.')
```

```python
# dependence plot for PaymentMethod
plt.figure(figsize=(8, 5))
shap.dependence_plot('PaymentMethod', test_shap_vals, X_test, interaction_index=None)

print('The Payment Method type that contributed most to churn prediction is electronic check.')
```
